```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('ggplot')
import numpy as np
```

```python
randomdata = np.random.RandomState(1)
X = 10 * randomdata.rand(50)
Y = 2 * X- 5 + randomdata.rand(50)
plt.scatter(X,Y)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X[:, np.newaxis],Y)  ## fit means learning from the data

## before fit it doesn't know about the relationship
## after fit , it knows about equation

xfit = np.linspace(0,10,1000)
yfit = model.predict(xfit[:, np.newaxis])

plt.scatter(X,Y)
plt.plot(xfit,yfit,color='green')
```
