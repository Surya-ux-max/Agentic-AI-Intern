```python
import spacy

nlp = spacy.load("en_core_web_sm")
```

```python
doc = nlp("Apple is launching a new AI product developed by Suryaprakash in Udumalapet.")
```

```python
for token in doc:
    print(token.text)
```

```python
for token in doc:
    print(token.text, token.pos_)
```

```python
for ent in doc.ents:
    print(ent.text, ent.label_)
```

## this model will find the what the text is , whether it is name or place

```python
doc = nlp(
    "Surya is a Engineering student."
)

for ent in doc.ents:
    print(ent.text, ent.label_)
```

```python
for token in doc:
    print(token.text, token.dep_)
```

## Other than Spacy , we can try Stanza , But spacy is much faster , high accuracy , easy than spanza



```python
import stanza

stanza.download("en")
```

```python
nlp = stanza.Pipeline("en")
```

```python
doc = nlp(
    "Apple is launching a new AI product."
)
```

```python
for sentence in doc.sentences:
    for word in sentence.words:
        print(word.text)
```

```python
for sentence in doc.sentences:
    for word in sentence.words:
        print(word.text, word.upos)
```
