### Defining the Features(X) and Target(Y) for my data

```python
X = df.drop(['Salary','Unnamed: 0'], axis=1)
 ## independent variable
Y = df['Salary']  ## dependent variable
```

```python
df.info()
```

### Splitting of dataset into training and testing part

X_train, X_test , Y_train , Y_test = train_test_split(X,Y)


```python
X_train, X_test , Y_train , Y_test = train_test_split(X,Y,test_size = 0.2,random_state = 42)

```

```python
X_train.shape
```

```python
Y_train.shape
```

```python
X_test.shape
```

```python
Y_test.shape
```

### Model Selection

```python
model = LinearRegression()
```

### Train Linear Regression Model

```python
model.fit(X_train,Y_train)
```

### Predicting the value

```python
predictedSalary = model.predict(X_test)
```

```python
predictedSalary
```

```python
errors = mean_squared_error(Y_test,predictedSalary)
print("Mean Squared Error:", errors)

r2 = r2_score(Y_test, predictedSalary)
print("R-squared:", r2)

mae = mean_absolute_error(Y_test, predictedSalary)
print("Mean Absolute Error:", mae)

mape = mean_absolute_percentage_error(Y_test, predictedSalary)
print("Mean Absolute Percentage Error:", mape)
```

```python
plt.scatter(X,Y,color='blue')
plt.plot(X_test,predictedSalary,color='red')
plt.title('Experience vs Salary')
plt.xlabel('Experience')
```
