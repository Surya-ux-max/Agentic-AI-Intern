### Understanding Embeddings

```python
from sentence_transformers import SentenceTransformer
model = SentenceTransformer('all-MiniLM-L6-v2')

chunks = ["AI is transforming education.",
              "RAG helps reduce hallucinations in LLM's.",
              "Embedding convert text into vectors for better understanding.",
              "Fine-tuning LLM's can improve performance on specific tasks.",
              "AI can personalize learning experiences for students.",
              "AI can automate grading and feedback for teachers.",
              "AI can assist in content creation for educational materials.",
              "AI can provide real-time language translation for students.",
              "Thalapathy Vijay is a popular Indian actor known for his versatile roles and charismatic screen presence.",
              "He has a massive fan following and is celebrated for his contributions to Tamil cinema.",
              "Vijay's films often feature a mix of action, romance, and social messages, making him a beloved figure in the industry.",
              "Vijay won 108 seats in assembly elections in Tamil Nadu.",
              "Vijay won in perambur as well as Trichy."
             ]

embedding = model.encode(chunks)
print(embedding)
print(embedding.shape)
```

### User Query

```python
query = "Where did Vijay won?"
print(query)
```

### Query Embedding

```python
Query_embedding = model.encode([query])
print(Query_embedding)
print(Query_embedding.shape)
```

### Install Scikit-learn

```python
! pip install scikit-learn
```

### Similarity Search

```python
from sklearn.metrics.pairwise import cosine_similarity

similarities = cosine_similarity([Query_embedding[0]], embedding)
print(similarities)
```

## Retrieve Best Matching Chunk

```python
best_match_index = similarities.argmax()
print("Best matching chunk:", chunks[best_match_index])
```

### Prompt Augmentation

```python
retrieved_chunk = chunks[best_match_index]
prompt = f"Based on the following information, answer the question: {retrieved_chunk}\nQuestion: {query}"
print(prompt)
```
